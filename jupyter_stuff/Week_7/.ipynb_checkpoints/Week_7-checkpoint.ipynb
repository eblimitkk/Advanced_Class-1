```python
from arcgis.gis import GIS
gis = GIS("https://www.arcgis.com","sspiegel_slu","Apples*200")
```

```python
import os

data_oldname = "stl_boundary_gjb.zip"
new_name = "stl_boundary_sss.zip"
os.rename(data_oldname,new_name)
```

```python
stl = gis.content.add({"title":"St. Louis Boundary - Steven Spiegel","tags":"St. Louis"},new_name)
```

```python
publish_stl = stl.publish()
```

```python
publish_stl
```

```python
publish_stl.access = 'public'
```

```python
publish_stl.access
```

```python
stl_map = gis.map('St. Louis')
```

```python
publish_stl.share(everyone=True,allow_members_to_edit=True)
```

```python
stl_map.add_layer(publish_stl)
```

```python
stl_map
```

```python
streets_old_name = 'streets_gjb.zip'
streets = 'streets_sss.zip'
os.rename(streets_old_name,streets)
```

```python
stl_streets = gis.content.add({"title":"St. Louis Boundary - Steven Spiegel","tags":"St. Louis streets"},streets)
```

```python
streets_published = stl_streets.publish()
```

```python
streets_published.share(everyone=True,allow_members_to_edit=True)
```

```python
stl_map.add_layer(streets_published)
```

```python
from arcgis.features import FeatureLayer
```

```python
fl = FeatureLayer(streets_published.layers[1].url, gis = gis)
```

```python
fl_query_layer = fl.query()
```

```python
sdf = fl_query_layer.df
```
