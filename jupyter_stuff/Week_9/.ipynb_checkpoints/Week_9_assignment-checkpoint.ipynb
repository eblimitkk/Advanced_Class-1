# Problem 1

```python
from arcgis.gis import GIS
import arcgis
from arcgis.features import SpatialDataFrame
```

```python
gis = GIS("http://slustl.maps.arcgis.com/home","sspiegel_slu","Apples*200")
```

```python
import pandas as pd
import os
```

```python
path = os.getcwd()
```

```python
file_list=[]
for file in os.listdir():
    if 'csv' in file:
        file_list.append(file)
```

```python
file_list
```

```python
earthquake = pd.read_csv(file_list[0])
```

```python
earthquake.head()
```

```python
earthquake_new = earthquake.drop(['nst','gap','dmin','net'],axis=1)
```

```python
earthquake_new.head()
```

```python
locations = earthquake_new.iloc[:,1:3].values.tolist()
```

```python
locations
```

```python
earthquake_sdf = SpatialDataFrame(earthquake_new,geometry=[arcgis.geometry.Geometry({'x':r[1], 'y':r[0], 
                    'spatialReference':{'wkid':4326}}) for r in locations])
```

```python
earthquake_sdf.head()
```

```python
earthquake_sdf=earthquake_sdf[earthquake_sdf.SHAPE.notna()].copy()
```

```python
earthquake_sdf.reset_index(inplace=True,drop=True)
```

```python
earthquake_list = []
```

```python
earthquake_sdf_low = earthquake_sdf[earthquake_sdf['mag'] < 2.5]
earthquake_list.append(earthquake_sdf_low)
```

```python
earthquake_sdf_lowmid = earthquake_sdf[(earthquake_sdf['mag']>2.5) & (earthquake_sdf['mag'] < 4)]
earthquake_list.append(earthquake_sdf_lowmid)
```

```python
earthquake_sdf_mid = earthquake_sdf[(earthquake_sdf['mag']>4) & (earthquake_sdf['mag'] < 5)]
earthquake_list.append(earthquake_sdf_mid)
```

```python
earthquake_sdf_high = earthquake_sdf[earthquake_sdf['mag'] > 5]
earthquake_list.append(earthquake_sdf_high)
```

```python
for earth in earthquake_list:
    earth.reset_index(inplace=True, drop=True)
```

```python
earthquake_sdf_high
```

```python
map1 = gis.map()
```

```python


classed_size_renderer = { "renderer": "ClassedSizeRenderer", "field_name":"mag"}

```

```python
eq = gis.content.import_data(earthquake_sdf_high, title='Magnitude Greater than 5')
```

```python
eq
```

```python
map1.add_layer(eq,classed_size_renderer)
```

```python
map1
```

```python
webmap_item_properties = {'title':'Earthquakes from Past Day, Greater than 5',
                         'snippet':'EARTHQUAKe!!!',
                         'tags':['day','mag>5','earthquake']}
```

```python
map1.save(webmap_item_properties)
```
